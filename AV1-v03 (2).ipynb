Quando os valores satisfizerem os requisitos para se formar um triângulo,  o código deve 'calcular' e mostrar na tela qual é o tipo de triângulo que pode ser montado (escaleno, isósceles ou equilátero).

```python
# Escreva aqui a resposta da questão 4
frase1 = str(input('Digite a primeira frase '))
frase2 = str(input("Digite a segunda frase "))
frase3 = str(input("Digite a terceira frase "))
frase = (frase1 +' '+ frase2 +' '+ frase3).split() 
contador = {}
for k in frase:
    contador.setdefault(k, 0)
    contador[k] += 1
for k, v in contador.items():
    print(f'A palavra "{k}" aparece {v} vez(es) na frase')
```
