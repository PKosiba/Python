1. Vamos criar uma tabela de preços com nome e o preço;
2. Inserir info nessa tabela

```python
# Remove o arquivo com o banco de dados SQLite (caso exista) e faz aqueles trâmites que vimos em outro documento
import os
os.remove("precos1.db") if os.path.exists("precos1.db") else None
# Importando o módulo de acesso ao SQLite
import sqlite3

# Cria uma conexão com o banco de dados. 
# Se o banco de dados não existir, ele é criado neste momento.
con = sqlite3.connect('precos1.db')

# Criando um cursor 
# (Um cursor permite percorrer todos os registros em um conjunto de dados)
cursor = con.cursor()

#criando uma tabela com 2 atributos: nome e preço
cursor.execute('''       
                create table precos(
                    nome text,
                    preco numeric)
                ''')

#inserindo elementos
#cursor.execute('''
#            insert into preços (nome, preço)
#            values(?, ?)
#            ''', ("Batata", "3.20"))

#outra forma de inserir elementos
cursor.execute('insert into precos (nome, preco) values("Batata", "3.77")')
                        
cursor.execute('''
            insert into precos (nome, preco)
            values(?, ?)
            ''', ("Pão", "1.20"))

cursor.execute('''
            insert into precos (nome, preco)
            values(?, ?)
            ''', ("Limao", "2.14"))
#acima fiz 3 inserts. Poderia usar o for? Sim, deveria.
# Grava a transação - sempre precisamos fazer o commit
con.commit()

#imprime os itens
cursor.execute('''select * from precos''')
for resultado in cursor.fetchall():
    print("Nome: {0:30s} Preço: {1:6.2f}".format(*resultado))
```
